```python
# 4) Write a Pandas code to filter out rows from a DataFrame named 'df' where column 'A' is less than 10 and column 'B' is greater than 20.

import pandas as pd
import numpy as np

df = pd.DataFrame({'A': [np.random.randint(0,50) for i in range(30)], 'B': [np.random.randint(0,50) for i in range(30)]})
print(df)

print(df[(df["A"] < 10) & (df["B"] > 20)])
```

```python
# 5) Write a Pandas code to group a DataFrame named 'df' by column 'A' and calculate the sum of column 'B' for each group.

import pandas as pd
import numpy as np

df = pd.DataFrame({'A': ["x", "y", "x", "z", "y", "y", "x", "z", "z"], 'B': [np.random.randint(0,50) for i in range(9)]})
print(df)

print(df.groupby("A").sum())
```

```python
# 6) Write a Pandas code to merge two DataFrames named 'df1' and 'df2' on the common column 'key'.

import pandas as pd
import numpy as np

a = list(set([np.random.randint(0, 20) for i in range(20)]))

df1 = pd.DataFrame({"A": a, "B": [np.random.randint(10, 40) for i in range(len(a))]})
print(df1)
print()
df2 = pd.DataFrame({"A": a, "c": [np.random.randint(20, 60) for i in range(len(a))]})
print(df2)
print()

df = pd.merge(df1, df2, on="A")
print(df)
```
